```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
minasGerais_harvest = pd.read_csv("/Users/sa21/Desktop/weather_api/data/csv/harvest_yearly_medians.csv")

minasGerais_harvest.head()
```

```python
minasGerais_harvest.describe()
```

The mean ans median numbers are close for the million_60kgs ans avg_unemp, the which indicate they no significant outliers, the distrubution is normal for the harvest data. Nonbear_mill trees is a litle bit left skewed.

```python
#drop nan values
minasGerais_harvest = minasGerais_harvest.dropna()
```

```python
#line plot
minasGerais_harvest[['million_60kgs_bag', 'nonbear_mill_trees', 'bear_mill_trees']].plot(
    figsize=(10, 6))
plt.xlabel("Year", fontsize=12)
plt.ylabel("Harvest and Tree Data", fontsize=12)
```

Nonbear mill trees and bear mills trees  fluctuate over the year. Million 60kgs bag look steady throughout the year but it's hard to tell.

```python
# Create the figure
plt.figure(figsize=(10, 6)) 
# set up bar plot 
sns.barplot(
    data = minasGerais_harvest,  
    x="year",
    y="million_60kgs_bag",
    palette="Blues" 
)
plt.title("Coffee Production in Minas Gerais", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Million 60kg Bags", fontsize=12)
plt.tight_layout()
plt.show()
```
